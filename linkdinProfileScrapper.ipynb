```python



def chrome(headless=False):
    # support to get response status and headers
    d = webdriver.DesiredCapabilities.CHROME
    d['loggingPrefs'] = {'performance': 'ALL'}
    opt = webdriver.ChromeOptions()
    if headless:
        opt.add_argument("--headless")
    opt.add_experimental_option('excludeSwitches', ['enable-logging'])
    opt.add_argument("--disable-popup-blocking")
    browser = webdriver.Chrome(executable_path=DRIVER_PATH, options=opt,desired_capabilities=d)
    browser.implicitly_wait(10)
    return browser

browser = chrome(True)
browser.get('https://www.linkedin.com/uas/login')
browser.implicitly_wait(3)
file = open('config.txt')
lines = file.readlines()
username = lines[0]
password = lines[1]

elementID = browser.find_element(By.ID, 'username')

elementID.send_keys(username)

elementID = browser.find_element(By.ID, 'password')
elementID.send_keys(password)

info = []

pages = 59
for i in range(1,pages):
    
    browser.get(#copy paste link here from linkedin search people which you want to scrape data)
    browser.implicitly_wait(1)
    
    
    try:
        
        searches =  browser.find_element(By.CLASS_NAME, 'search-results-container')
        t = searches.find_element(By.TAG_NAME, 'ul')
        people = t.find_elements(By.TAG_NAME, 'li')
    
    except:
        continue
        
        
    for j in range(len(people)):
        
        searches =  browser.find_element(By.CLASS_NAME, 'search-results-container')
        t = searches.find_element(By.TAG_NAME, 'ul')
        people = t.find_elements(By.TAG_NAME, 'li')
        
        time.sleep(2)
        people[j].click()
        time.sleep(2)
        p = False
        try:
            pop = browser.find_element(By.XPATH, '/html/body/div[3]/div/div/button').click()
            p = True
            time.sleep(3)
        except:
            p = False


        if (p == False):



            src = browser.page_source
            soup = BeautifulSoup(src, 'lxml')

            # Get Name of the person


            try:
                name_div = soup.find('div', {'class': 'mt2 relative'})
                first_last_name = name_div.find('h1').get_text().strip()
            except:
                first_last_name = None

            # Get Talks about section info
            try:
                talksAbout_tag = name_div.find('div', {'class': 'text-body-small t-black--light break-words mt2'})
                talksAbout = talksAbout_tag.find('span').get_text().strip()
            except:
                talksAbout = None
            # Get Location of the Person
            try:
                location_tag = name_div.find('div', {'class': 'pv-text-details__left-panel mt2'})
                location = location_tag.find('span').get_text().strip()
            except:
                location = None


            # Get Company Link of the Person
            try:
                exp_section = soup.find('section', {'id':'ember66'})
                exp_section = exp_section.find('ul')
                li_tags = exp_section.find('div')
                a_tags = li_tags.find('a')

                company_link = a_tags['href']
                company_link = 'https://www.linkedin.com/' + company_link
            except:
                company_link = None

            # Get Job Title of the Person
            try:
                job_title = name_div.find('div',{'class':'text-body-medium break-words'}).get_text().strip()

            except:
                job_title = None

            # Get Company Name of the Person
            try:
                company_name = li_tags.find('p', {'class': 'pv-entity__secondary-title t-14 t-black t-normal'}).get_text().strip()
            except:
                company_name = None

            try:
                name_div = soup.find('div', {'class': 'mt2 relative'})
                contact_page = name_div.find('div', {'class': 'pv-text-details__left-panel mt2'})
                con = contact_page.find('a')['href']
                a = con.split("/")
                b = a[:3]
                c = "/".join(b)
                contacts = "linkedin.com"+c
            except:
                contacts = None

            if(first_last_name != None):
                info.append([first_last_name, job_title, talksAbout, location, contacts])
                time.sleep(1)
                browser.back()
                time.sleep(2)
            time.sleep(3)



```
